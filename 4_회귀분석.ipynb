공공체육시설 데이터 불러오기

```python
df_2011 = pd.read_excel('/content/drive/MyDrive/project/0. 공공체육시설/서울시_공공체육시설_통계_2011년.xls')
df_2011
```

```python
df_2011 = df_2011[['자치구', '공공체육시설']]
df_2011
```

```python
df_2011.drop(df_2011.index[0:4], inplace=True)
df_2011
```

```python
df_2011 = df_2011.reset_index()
df_2011.drop(['index'], axis=1, inplace=True)
```

```python
df_2011
```

```python
df_2011.drop(df_2011.index[25], inplace=True)
df_2011
```

```python
# 비만도 데이터 불러오기

BMI_2011 = pd.read_csv('/content/drive/MyDrive/project/3. 비만도 데이터/BMI_2011.csv', index_col='Unnamed: 0')
BMI_2011
```

```python
# BMI 이상

BMI_2011 = BMI_2011[['자치구', '이상']]
BMI_2011
```

```python
merge_2011 = pd.merge(df_2011, BMI_2011)
merge_2011
```

```python
from statsmodels.formula.api import ols


ols("이상 ~ 공공체육시설", data=merge_2011).fit().summary()
```

```python
merge_2011['공공체육시설'] = merge_2011['공공체육시설'].astype(int)
```

```python
# 2. matplotlib의 폰트를 Nanum 폰트로 지정
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic') 
```

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
 
#pyplot과 seaborn으로 회귀그래프 
plt.rcParams["figure.figsize"] = (12, 6)
sns.regplot(x='공공체육시설', y='이상', data=merge_2011)
plt.ylabel('BMI 이상')
plt.xlim(merge_2011['공공체육시설'].min()-1, merge_2011['공공체육시설'].max()+1)
plt.grid()
plt.show()
```

```python
z = np.polyfit(merge_2011['공공체육시설'], merge_2011['공공체육시설'], 1)
f = np.poly1d(z)

print(z[0], z[1])
print(f(1))
```

```python
plt.rcParams["figure.figsize"] = (12, 6)
sns.regplot(x='공공체육시설', y='이상', data=merge_2011)
plt.xlim(merge_2011['공공체육시설'].min()-1, merge_2011['공공체육시설'].max()+1)
plt.ylabel('BMI 이상')
plt.text(5, 100, "y = %fx + %f" %(z[0], z[1]), color="#005599")
plt.grid()
plt.show()
```
